***DATA CLEANING***

```python
# change the datatype of date column
df['Date']=pd.to_datetime(df['Date'],format='%Y-%m-%d')
```

```python
#check the how many columns are categorical 
cat_var = df.select_dtypes(include=['O'])
cat_var_count = len(cat_var.columns)
print("Number of categorical columns =", cat_var_count)
```

```python
#check the how many columns are Numerical 
num_var = df.select_dtypes(include=['float64'])
num_var_count = len(num_var.columns)
print("Number of numerical columns =", num_var_count)
```

```python
#check the null values in data frame
miss_null=df.isnull().mean()*100
miss_null[miss_null>0]
```

```python
#check in cat_column how many null values are their
cat_var=df.select_dtypes(include=['O'])
miss_car=cat_var.isnull().mean()*100
miss_car[miss_car>0]
```

```python
# Remove the space between column name and add underscore
df.columns = df.columns.str.replace(" ", "_")
df.columns
```

```python
#drop columns where null value is greater than 80%
del_cat=['Reason_for_cancelling_by_Customer','Driver_Cancellation_Reason','Incomplete_Rides_Reason']
df.drop(columns=del_cat,axis=1,inplace=True)
```

```python
#check in numerical_column how many null values are their
num_var=df.select_dtypes(include=['float64'])
miss_num=num_var.isnull().mean()*100
miss_num[miss_num>0]
```

```python
#drop columns where null value is greater than 80%
del_num=['Cancelled_Rides_by_Driver','Cancelled_Rides_by_Customer','Incomplete_Rides']
df.drop(columns=del_num,axis=1,inplace=True)
```

```python
cat_var=df.select_dtypes(include=['O'])
miss_cat=cat_var.isnull().mean()*100
miss_cat[miss_cat>0]
```

```python
#filling null values in categorical columns by using fillna
for var in cat_var:
    df[var].fillna(df[var].mode()[0],inplace=True)
    print(var,'=',df[var].mode()[0])
```

```python
num_var=df.select_dtypes(include=['float64'])
miss_num=num_var.isnull().mean()*100
miss_num[miss_num>0]
```

```python
#filling null values in numerical columns by using fillna
for var in num_var:
    df[var].fillna(df[var].mean().round(),inplace=True)
    print(var,'=',df[var].mean().round())
```

```python
# Again check the null values in df
df.isnull().mean()*100
```

```python
# calculate the sum of booking value where booking is completed group by on the behalf of vehicle type
fc=df[df['Booking_Status']=='Completed']
fc.groupby('Vehicle_Type')['Booking_Value'].sum().reset_index().sort_values('Booking_Value',ascending=False)
```

```python
#calculate the sum of booking value where booking is Incompleted group by on the behalf of vehicle type
fc=df[df['Booking_Status']=='Incomplete']
fc.groupby('Vehicle_Type')['Booking_Value'].sum().reset_index().sort_values('Booking_Value',ascending=False)
```

```python
#calculate the sum of booking value where booking is Incompleted group by on the behalf of vehicle type
fc=df[df['Booking_Status']=='No Driver Found']
fc.groupby('Vehicle_Type')['Booking_Value'].sum().reset_index().sort_values('Booking_Value',ascending=False)
```

```python
# Filter for customer-canceled rides
fc = df[df['Booking_Status'] == 'Cancelled by Customer']

# Group by Customer_ID, count number of cancellations, sort descending, get top 5
top_5_cancel_customers = (
    fc.groupby('Customer_ID')
      .size()
      .reset_index(name='Cancel_Count')
      .sort_values('Cancel_Count', ascending=False)
      .head(5)
)

top_5_cancel_customers
```

```python
# Growth_Percentage
fc = df[df['Booking_Status'] == 'Completed']
total_sum = fc['Booking_Value'].sum()

Month_wise_sale = (
    fc.groupby(['Month_Name', 'Month_Number'])['Booking_Value']
    .sum()
    .reset_index()
    .sort_values('Month_Number', ascending=True)
)

Month_wise_sale['Growth_Percentage'] = (Month_wise_sale['Booking_Value'] / total_sum) * 100

Month_wise_sale
```
